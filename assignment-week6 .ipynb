```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets
import bqplot
import numpy as np
```

```python
f = pd.read_csv("C:/Users/sjnnt/is590dv/building_inventory.csv")
f
```

# Heat Map

```python
f_g = f['Square Footage'].groupby([f['Agency Name'],f['Congress Dist']]).sum().reset_index(name = 'Total Square Footage')
f_g.head(10)
```

```python
sf = f_g['Total Square Footage'].unique()
an = f_g['Agency Name'].unique()

cd = f_g['Congress Dist'].unique()
cd.sort()
cd
```

Since the np.hist2d is only expecting numerical data, I manually construct a matrix for the weight values.

```python
ft_sum = np.zeros([len(an),len(cd)])
for i,x in enumerate(an):
    for j,y in enumerate(cd):
        mask = ((f_g['Congress Dist'] == y) & (f_g['Agency Name'] == x))
        ft_sum[i,j] = f_g['Total Square Footage'][mask].sum()
ft_sum
```

```python
ft_sum[ft_sum<=0] = np.nan
ft_sum=ft_sum.T
ft_sum
```

```python

x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()
col_sc = bqplot.ColorScale(scheme='RdPu',
                           min = np.nanmin(ft_sum),
                           max = np.nanmax(ft_sum))

x_ax = bqplot.Axis(scale=x_sc, label='Agency Name',label_offset = '20px',
                   tick_rotate = 45,tick_style = {'tick_offset':'100px','text_anchor':'top'})
y_ax = bqplot.Axis(scale=y_sc, label='Congress Dist',orientation='vertical')
c_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')
```

```python
heat_map = bqplot.GridHeatMap(color = ft_sum,
                              row =cd,column = an,
                              scales={'color':col_sc, 'row':y_sc, 'column':x_sc},
                             interactions={'click':'select'}, 
                             anchor_style={'fill':'blue'})

fig = bqplot.Figure(marks = [heat_map], axes = [c_ax, y_ax, x_ax])
fig
```

# Line plot

```python
x_scl = bqplot.LinearScale()
y_scl = bqplot.LinearScale()

axl_x = bqplot.Axis(label='Year', scale=x_scl )
axl_y = bqplot.Axis( label='Total square footage', scale=y_scl, 
                    orientation='vertical',side ='left')
```

Since there are some dirty data in column "Year Acquired", I cleaned the data.

```python
f_l = f['Square Footage'].groupby([f['Agency Name'],f['Congress Dist'],f['Year Acquired']]).sum().reset_index(name = 'Total Square Footage')
f_l = f_l[f_l['Year Acquired'] > 0]
f_l.head(10)
```

```python
print(len(an),len(cd))
print(ft_sum.shape)
```

```python
i,j = 0,11
mask_l = (f_l['Agency Name'].values == an[j]) & (f_l['Congress Dist'].values == cd[i])
```

```python
line_plot = bqplot.Lines(x = f_l['Year Acquired'][mask_l],
                         y = f_l['Total Square Footage'][mask_l],
                        scales = {'x':x_scl, 'y':y_scl})
fig = bqplot.Figure(marks = [line_plot], axes=[axl_x, axl_y])
fig
```

# put them together

```python
def get_data_value(change):
    # to make sure we only support single selections
    if len(change['owner'].selected) == 1: # *only* 1 selection
        i,j = change['owner'].selected[0]
        v = ft_sum[i,j]
        mySelectedLabel.value = 'Total Square Footages = ' + str(v)
        # UPDATE LINE PLOT TOO
        agen_n =an[j]
        cong_d = cd[i]
        mask_l = (f_l['Agency Name'].values == an[j]) & (f_l['Congress Dist'].values == cd[i])

        line_plot.x = f_l['Year Acquired'][mask_l]
        line_plot.y = f_l['Total Square Footage'][mask_l]
          
# make sure we "observe" for a change in our heatmap (traitlets)
heat_map.observe(get_data_value, 'selected')
```

```python
fig_heatmap = bqplot.Figure(marks=[heat_map], axes=[c_ax, y_ax, x_ax])
fig_lines = bqplot.Figure(marks=[line_plot], axes=[axl_x, axl_y])
```

```python
mySelectedLabel = ipywidgets.Label()
fig_heatmap.layout.min_width='600px'
fig_lines.layout.min_width = '400px'

plots = ipywidgets.HBox([fig_heatmap, fig_lines])
myDashboard = ipywidgets.VBox([mySelectedLabel, plots])
myDashboard
```

# Can you keep the x and y ranges static on the line plot?
